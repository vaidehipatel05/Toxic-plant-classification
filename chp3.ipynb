```python
import os

# Function to calculate size and count of images for each class in the dataset
import os
from tabulate import tabulate

# Function to calculate size and count of images for each class in the dataset
def calculate_classwise_size(root_folder_path):
    class_sizes = []
    for subfolder in os.listdir(root_folder_path):
        subfolder_path = os.path.join(root_folder_path, subfolder)
        if os.path.isdir(subfolder_path):
            class_size_bytes = 0
            image_count = 0
            for subdir, _, files in os.walk(subfolder_path):
                for file in files:
                    file_path = os.path.join(subdir, file)
                    class_size_bytes += os.path.getsize(file_path)
                    image_count += 1
            class_sizes.append([subfolder, class_size_bytes / (1024 * 1024), image_count])
    return class_sizes


# Function to calculate size for the entire dataset
def calculate_total_size(root_folder_path):
    total_size_bytes = 0
    for subfolder in os.listdir(root_folder_path):
        subfolder_path = os.path.join(root_folder_path, subfolder)
        if os.path.isdir(subfolder_path):
            for subdir, _, files in os.walk(subfolder_path):
                for file in files:
                    file_path = os.path.join(subdir, file)
                    total_size_bytes += os.path.getsize(file_path)
    total_size_mb = total_size_bytes / (1024 * 1024) # Convert bytes to MB
    return total_size_mb

# Provide the path to your root image dataset folder
root_folder_path = "/Users/vaidehipatel/Documents/Course Sem 2/GWAR/Toxic_Plants_Dataset_resized"

# Calculate size and count of images for each class
class_sizes = calculate_classwise_size(root_folder_path)

# Calculate total size
total_size = calculate_total_size(root_folder_path)

# Print total size
print("\n------ Total Data Size ------")
print(f"Total Size (MB): {total_size:.2f}")

# Provide the path to your root image dataset folder
root_folder_path = "/Users/vaidehipatel/Documents/Course Sem 2/GWAR/Toxic_Plants_Dataset_resized"

# Calculate size and count of images for each class
class_sizes = calculate_classwise_size(root_folder_path)

# Print as table
print("\n------ Class-wise Size and Image Count ------")
print(tabulate(class_sizes, headers=['Class Name', 'Size (MB)', 'Image Count'], tablefmt='grid'))

```

```python
import os

# Function to calculate size and count for each class in the dataset
def calculate_classwise_size_and_count(root_folder_path):
    class_info = {}
    for subfolder in os.listdir(root_folder_path):
        subfolder_path = os.path.join(root_folder_path, subfolder)
        if os.path.isdir(subfolder_path):
            class_size_bytes = 0
            file_count = 0
            for subdir, _, files in os.walk(subfolder_path):
                for file in files:
                    file_path = os.path.join(subdir, file)
                    class_size_bytes += os.path.getsize(file_path)
                    file_count += 1
            class_info[subfolder] = {"size_mb": class_size_bytes / (1024 * 1024), "file_count": file_count}
    return class_info

# Function to calculate size for the entire dataset
def calculate_total_size(root_folder_path):
    total_size_bytes = 0
    total_file_count = 0
    for subfolder in os.listdir(root_folder_path):
        subfolder_path = os.path.join(root_folder_path, subfolder)
        if os.path.isdir(subfolder_path):
            for subdir, _, files in os.walk(subfolder_path):
                for file in files:
                    file_path = os.path.join(subdir, file)
                    total_size_bytes += os.path.getsize(file_path)
                    total_file_count += 1
    total_size_mb = total_size_bytes / (1024 * 1024) # Convert bytes to MB
    return {"size_mb": total_size_mb, "file_count": total_file_count}

# Provide the path to your root image dataset folder
root_folder_path = "/Users/vaidehipatel/Documents/Course Sem 2/GWAR/Toxic_Plants_Dataset_resized"

# Calculate size and count for each class
class_info = calculate_classwise_size_and_count(root_folder_path)

# Calculate total size and count
total_info = calculate_total_size(root_folder_path)

# Print total size and count
print("\n------ Total Data Size ------")
print(f"Total Size (MB): {total_info['size_mb']:.2f}")
print(f"Total Files: 6242")

# Print size and count for each class
print("\n------ Class-wise Size and Count ------")
for class_name, info in class_info.items():
    print(f"{class_name}:")
    print(f"  - Size (MB): {info['size_mb']:.2f}")
    print(f"  - File Count: {info['file_count']}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
plants = ['Rosary Pea', 'Cicuta', 'English Yew', 'Manchineel', 'Angel Trumpet', 
          'Castor Bean', 'Datura', 'Foxglove', 'White Snakeroot', 'Lily of the valley']
resizing = [7.27, 16.49, 21.44, 9.90, 19.92, 16.45, 18.60, 22.31, 19.75, 18.30]
denoising = [7.89, 24.40, 25.08, 11.21, 22.63, 18.65, 21.68, 25.19, 22.76, 20.41]

# Set the width of the bars
bar_width = 0.35

# Set the position of the bars on the x-axis
r1 = np.arange(len(plants))
r2 = [x + bar_width for x in r1]

# Plot
plt.figure(figsize=(12, 6))

plt.bar(r1, resizing, color='blue', width=bar_width, edgecolor='grey', label='Resizing')
plt.bar(r2, denoising, color='skyblue', width=bar_width, edgecolor='grey', label='Denoising')

plt.xlabel('Plant', fontweight='bold')
plt.ylabel('Size (in MB)', fontweight='bold')
plt.title('Resizing vs Denoising Size Comparison', fontweight='bold')
plt.xticks([r + bar_width/2 for r in range(len(plants))], plants, rotation=45, ha='right')
plt.legend()

plt.tight_layout()
plt.show()
```
